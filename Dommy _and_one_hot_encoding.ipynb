```python
import pandas as pd 
import numpy as np 
from sklearn import linear_model 
```

```python
df = pd.read_csv("C:\\Users\\nithi\\Downloads\\carprices.csv")
df
```

```python
df.columns.str.replace(" ","_")
```

```python
df.columns = df.columns.str.replace(" ","_")
df
```

```python
dummies = pd.get_dummies(df.Car_Model)
dummies
```

```python
change =dummies.astype(int)
change
```

```python
final = pd.concat([df,change],axis="columns")
final
```

```python
fin = final.drop(["Car_Model","Mercedez Benz C class"],axis="columns")
fin
```

```python
x=fin.drop(["Sell_Price($)"],axis="columns")
x
```

```python
y = fin["Sell_Price($)"]	
y
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(x,y)
```

```python
import warnings
warnings.filterwarnings("ignore", message="X does not have valid feature names")
```

```python
model.predict([[45000,4,0,0]])
```

**Price of mercedez benz that is 4 yr old with mileage 45000**


```python
model.predict([[45000,4,0,0]])
```

**Price of BMW X5 that is 7 yr old with mileage 86000**

```python
model.predict([[86000,7,0,1]])
```

**Predicted Accurancy**

```python
model.score(x,y)
```
